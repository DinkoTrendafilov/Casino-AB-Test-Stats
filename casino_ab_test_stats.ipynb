```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
successes_a = int(input("Enter success of group A: "))
visitors_a = int(input("Enter total visitors of group A: "))
successes_b = int(input("Enter success of group B: "))
visitors_b = int(input("Enter total visitors of group B: "))
```

```python
p_a = successes_a / visitors_a
p_b = successes_b / visitors_b
diff = p_b - p_a

print(f"Results:")
print(f"  Group A: {successes_a}/{visitors_a} = {p_a:.2%}")
print(f"  Group B: {successes_b}/{visitors_b} = {p_b:.2%}")
print(f"  Difference (B-A): {diff:+.2%}")
```

```python
count = [successes_a, successes_b]
nobs = [visitors_a, visitors_b]
z_score, p_value = proportions_ztest(count, nobs)

alpha = 0.05

print(f"Statistical Test:")
print(f"  z-stats = {z_score:.4f}")
print(f"  p-value = {p_value:.4f} --- {p_value:.4%}")
```

```python
if p_value < alpha:
    print(f"SIGNIFICANT! (p < {alpha})")
    if diff > 0:
        print(f"     → Group B WINS by {diff * 100:.1f} percentage points!")
    elif diff < 0:
        print(f"     → Group A WINS by {abs(diff) * 100:.1f} percentage points!")
    else:
        print(f"     → No difference (identical rates)")
else:
    print(f"NOT significant! (p > {alpha})")
    print(f"     → No clear winner - difference could be due to chance")
```
